```python
# You code here (Please add comments in the code):
# Access to school records would be necessary to gather real-world data on student attendance and academic performance for 1000 students which I do not have.
# So I have just taken random values and written a code according to that. We can use this basic code if we have access to the real world data on school.
import random
import csv

num_students = 1000
csv_path = 'student_data.csv' # I have defined a path where it saves the data into the CSV file

# Create a CSV file to store the artificial data
with open(csv_path, "w", newline="") as csvfile:
    fieldnames = ["Student_ID", "Attendance_Days", "GPA"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader() # Write the header row
   # Create artificial data for 1000 students
    for student_id in range(1, num_students + 1):
        # Simulate attendance data (between 120 and 180 days in a school year)
        attendance_days = random.randint(120, 180)

        # Simulate GPA data (between 2.0 and 4.0)
        gpa = round(random.uniform(2.0, 4.0), 2)

        # Write the data to the CSV file
        writer.writerow({"Student_ID": student_id, "Attendance_Days": attendance_days, "GPA": gpa})

print(f"Artificial student data collected and saved to {csv_path}.")

# Read and print the first few rows of the generated data for verification
with open(csv_path, "r") as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)  # Read the header row
    print(header)  # Print the header row
    for i in range(1000):  # Print the first 1000 rows of data as an example
        row = next(reader)
        print(row)
```

Question 3 (10 points): Write python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/), or ACM Digital Libraries (https://dl.acm.org/) with the keyword "information retrieval". The articles should be published in the last 10 years (2013-2023).

The following information of the article needs to be collected:

(1) Title

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
# You code here (Please add comments in the code):
import requests
from bs4 import BeautifulSoup

# Function to scrape and print articles from a given URL
def scrape_articles(url):
    response = requests.get(url)  # Send a GET request to the URL
    soup = BeautifulSoup(response.text, 'html.parser')   # Use BeautifulSoup to extract the page's HTML content.
    articles = soup.find_all('div', class_='gs_ri')  # Locate and gather data for each article on the page.

  # Loop through the articles and extract the desired information
    for article in articles:
        title = article.find('h3', class_='gs_rt').text
        venue = article.find('div', class_='gs_a').text.split(' - ')[-1]
        year = article.find('div', class_='gs_a').text.split(' - ')[0]
        authors = article.find('div', class_='gs_a').text.split(' - ')[1]
        abstract = article.find('div', class_='gs_rs').text

        print("Title:", title)
        print("Venue:", venue)
        print("Year:", year)
        print("Authors:", authors)
        print("Abstract:", abstract)
        print("\n---\n")

# Loop through plenty of pages of results to get 1,000 articles.
articles_per_page = 10  # A page of Google Scholar normally shows 10 items.
total_articles = 1000

for page in range(0, total_articles, articles_per_page):
  # URL of Google Scholar search results
    page_url = f"https://scholar.google.com/scholar?hl=en&as_sdt=0%2C5&q=information+retrieval&as_ylo=2013&as_yhi=2023&start={page}"
    scrape_articles(page_url)
```

Question 4 (10 points): Write python code to collect 1000 posts from Twitter, or Facebook, or Instagram. You can either use hashtags, keywords, user_name, user_id, or other information to collect the data.

The following information needs to be collected:

(1) User_name

(2) Posted time

(3) Text

```python
import requests
from bs4 import BeautifulSoup

# Define the Instagram profile or hashtag URL to scrape
url = "https://www.instagram.com/explore/tags/your_hashtag/"

# Send a GET request to the Instagram page
response = requests.get(url)

# Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find and extract information for each post on the page
posts = soup.find_all('div', class_='v1Nh3')

# Loop through the posts and extract the desired information
for post in posts:
    # Extract post details such as user_name, posted_time, and text
    user_name = post.find('a', href=True)['href'].split('/')[1]  # Extract username from href
    posted_time = post.find('time')['datetime']
    text = post.find('img')['alt']  # Assuming the alt text contains the post caption

    print("User_name:", user_name)
    print("Posted_time:", posted_time)
    print("Text:", text)
    print("\n---\n")
```
